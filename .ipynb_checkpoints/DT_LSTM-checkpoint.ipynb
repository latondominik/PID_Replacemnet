```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import time

# For LSTM model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
from keras.models import load_model


from plotly.subplots import make_subplots
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
# Load training data

DT_1 = pd.read_csv('DT_garnek_1th_try.csv')
DT_2 = pd.read_csv('DT_garnek_1th_try_continue.csv')
DT_3 = pd.read_csv('DT_garnek_1th_try_continue_2.csv')
DT_4 = pd.read_csv('DT_garnek_1th_try_continue_3.csv')

# train = pd.concat([DT_1, DT_2, DT_3])
train = pd.concat([DT_1, DT_2, DT_3, DT_4])

train = train.reset_index()
train['temperature'] = train['temperature'].apply(lambda x: 100 if x > 100 else x)
```

```python
fig = make_subplots(rows=2, cols = 1)

fig.add_trace(
    go.Scatter(y=train["temperature"], name="Temperatura zmierzona"), row=1, col=1
)

fig.add_trace(
    go.Scatter(y=train["control"], name="Sterowanie"), row=2, col=1
)

fig.update_layout(height=900, width=1100)
fig.show()
```

```python
window_to_avg = 10
train['smooth_temperature'] = train['temperature'].rolling(window_to_avg).sum()/window_to_avg
# train['smooth_temperature'] = train['temperature']
```

```python
fig = make_subplots(rows=2, cols = 1)

fig.add_trace(
    go.Scatter(y=train["smooth_temperature"], name="Uśredniona"), row=1, col=1
)

fig.add_trace(
    go.Scatter(y=train["temperature"], name="Zmierzona"), row=2, col=1
)

fig.update_layout(height=900, width=1100)
fig.show()
```

```python
# Scale features
s1 = MinMaxScaler(feature_range=(-1,1))
Xs = s1.fit_transform(train[['smooth_temperature','control']])

# Scale predicted value
s2 = MinMaxScaler(feature_range=(-1,1))
Ys = s2.fit_transform(train[['smooth_temperature']])
```

```python
# Each time step uses last 'window' to predict the next change
window = 30
X = []
Y = []
for i in range(window,len(Xs)):
    X.append(Xs[i-window:i,:])
    Y.append(Ys[i])

# Reshape data to format accepted by LSTM
X, Y = np.array(X), np.array(Y)
X = X[window:]
Y = Y[window:]
# create and train LSTM model

# Initialize LSTM model
model = Sequential()

model.add(LSTM(units=40, return_sequences=True, \
          input_shape=(X.shape[1],X.shape[2])))
model.add(Dropout(0.4))
model.add(LSTM(units=40, return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(units=40))
model.add(Dropout(0.4))
model.add(Dense(units=1))
model.compile(optimizer = 'adam', loss = 'mean_squared_error',
              metrics = ['accuracy'])

# Allow for early exit
es = EarlyStopping(monitor='loss',mode='min',verbose=1,patience=10)
```

```python
# Fit (and time) LSTM model
t0 = time.time()
history = model.fit(X, Y, epochs = 20, batch_size = 360, callbacks=[es], verbose=1)
t1 = time.time()
print('Runtime: %.2f s' %(t1-t0))
```

```python
# Plot loss
plt.figure(figsize=(8,4))
plt.semilogy(history.history['loss'])
plt.xlabel('epoch'); plt.ylabel('loss')
plt.savefig('tclab_loss.png')
model.save('model_6.keras')
```

```python
# Verify the fit of the model
Yp = model.predict(X)

# un-scale outputs
Yu = s2.inverse_transform(Yp)
Ym = s2.inverse_transform(Y)

plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(Yu,'r-',label='LSTM')
plt.plot(Ym,'k--',label='Measured')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.subplot(2,1,2)
plt.plot(train['control'],label='heater (%)')
plt.legend()
plt.xlabel('Time (sec)'); plt.ylabel('Heater')
# plt.savefig('garnek_fit.png')
plt.show()
```

```python
# Load model
model = load_model('model_6.keras')
window = 30
window_to_avg = 10
# Load training data
# test = pd.read_csv('PID_garnek_6th_try_en.csv')
test = pd.read_csv('DT_garnek_2th_try.csv')
test.drop(index = 0, inplace= True)
test['temperature'] = test['temperature'].apply(lambda x: 100 if x > 100 else x)
test['smooth_temperature'] = test['temperature'].rolling(window_to_avg).sum()/window_to_avg
test['control'] = test['control'].rolling(window_to_avg).sum()/window_to_avg
# test['smooth_temperature'] = test['temperature']
Xt = test[['smooth_temperature','control']].values
Yt = test[['smooth_temperature']].values

Xts = s1.transform(Xt)
Yts = s2.transform(Yt)


Xts = Xts[window:]
Yts = Yts[window:]

Xti = []
Yti = []
for i in range(window,len(Xts)):
    Xti.append(Xts[i-window:i,:])
    Yti.append(Yts[i])

# Reshape data to format accepted by LSTM
Xti, Yti = np.array(Xti), np.array(Yti)
```

```python
# Verify the fit of the model
Ytp = model.predict(Xti)
# un-scale outputs
Ytu = s2.inverse_transform(Ytp)
Ytm = s2.inverse_transform(Yti)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(Ytu,'r-',label='LSTM Predicted')
plt.plot(Ytm,'k--',label='Measured')
plt.legend()
plt.ylabel('Temperature (°C)')
plt.subplot(2,1,2)
# plt.plot(Ytm,'k--',label='Measured')
plt.plot(test['control'],'b-',label='Heater')
plt.xlabel('Time (sec)'); plt.ylabel('Heater (%)')
plt.legend()
plt.show()
```

```python
# Using predicted values to predict next step

Xtsq = Xts.copy()
for i in range(window,len(Xtsq)):
    Xin = Xtsq[i-window:i].reshape((1, window, 2))
    Xtsq[i][0] = model.predict(Xin)
    Yti[i-window] = Xtsq[i][0]

#Ytu = (Yti - s2.min_[0])/s2.scale_[0]
Ytu = s2.inverse_transform(Yti)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(Ytu,'r-',label='LSTM Predicted')
plt.plot(Ytm,'k--',label='Measured')
plt.legend()
plt.ylabel('Temperature (°C)')
plt.subplot(2,1,2)
plt.plot(test['control'][window:],'b-',label='Heater')
plt.xlabel('Time (sec)'); plt.ylabel('Heater (%)')
plt.legend()
plt.savefig('garnek_forecast.png')
plt.show()
```
